# SEABORN â€” Statistical Data Visualization

## What is Seaborn?

- **Seaborn** is a Python library built on top of **Matplotlib**.
- It makes **attractive, informative, and statistical plots** with fewer lines of code.
- Integrates perfectly with **Pandas DataFrames**.

## ðŸ“¦ Importing Seaborn



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Sample Dataset

```python
# Load sample dataset from Seaborn
tips = sns.load_dataset('tips')
tips.head()
```

 ## 1. Bar Plot â€” Compare Means
 Shows average value (with error bars). Good for comparing group means.



```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Average Total Bill per Day")
plt.show()
```

## 2. Line Plot â€” Trend Over X-axis
Great for time series or progression-style data.

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot")
plt.show()
```

 ## 3. Scatter Plot â€” Relationship Between Two Variables
 Use to detect correlation, clusters, or outliers.



```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Bill vs Tip")
plt.show()
```

## 4. Histogram â€” Distribution of a Single Variable
Shows how data is spread. kde=True adds a smooth density curve.

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

## 5. Box Plot â€” Statistical Summary
Helps detect outliers, medians, and data spread

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill per Day")
plt.show()
```

## 6. Violin Plot â€” Distribution + Density
Combines boxplot + density curve. More informative for skewed data.

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot")
plt.show()
```

## 7. Heatmap â€” Correlation Matrix or Grid
 Visualizes pairwise correlation between numeric features.

```python
correlation = tips.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## 8. Pair Plot â€” All-in-One Relationships
 Plots scatterplots & histograms for all features â€” powerful for exploratory data analysis (EDA).

```python
sns.pairplot(tips, hue='sex')
```

## 9. Styling and Themes
Themes: white, whitegrid, dark, darkgrid, ticks

```python
sns.set_style("whitegrid")
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Styled Plot with Whitegrid Theme")
plt.show()
```

## Summary of Seaborn Plots

| Plot Type    | Function            | Best For                        |
| ------------ | ------------------- | ------------------------------- |
| Bar Plot     | `sns.barplot()`     | Comparing group means           |
| Line Plot    | `sns.lineplot()`    | Showing trends or time series   |
| Scatter Plot | `sns.scatterplot()` | Correlation between variables   |
| Histogram    | `sns.histplot()`    | Frequency distribution          |
| Box Plot     | `sns.boxplot()`     | Outliers, median, IQR           |
| Violin Plot  | `sns.violinplot()`  | Shape + summary of distribution |
| Heatmap      | `sns.heatmap()`     | Correlation, matrix data        |
| Pair Plot    | `sns.pairplot()`    | Full EDA in one line            |


Seaborn makes your data visualizations more elegant, powerful, and insightful â€” especially for statistical analysis and storytelling.
