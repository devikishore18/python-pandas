## Importing

```python
import matplotlib.pyplot as plt #Conventionally, we use plt as the alias.
```

## Line Plot

**Use when**: Showing trends, progression, or changes over time or sequence.

**Explanation:**
Connects data points with lines.
Great for time series, stock prices, temperature over days, etc.

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

## Bar Chart

**Use when**: Comparing categorical data like sales per product or marks per subject.

**Explanation:**
Vertical bars represent values for different categories.
Use for side-by-side comparisons.



```python
x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]

plt.bar(x, y, color='teal')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Pie Chart

**Use when:** Displaying proportions or percentages of a whole.

**Explanation:**
Circular chart divided into slices.
Good for market share, survey results, etc.
Avoid with too many categories.





```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Programming Language Usage")
plt.axis('equal')
plt.show()
```

## Scatter Plot


**Use when:** Analyzing relationships/correlations between two continuous variables.

**Explanation:**
Each point represents one observation.
Useful for spotting clusters, outliers, and correlations.

```python
x = [5, 7, 8, 7, 2, 17, 2]
y = [99, 86, 87, 88, 100, 86, 103]

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.grid()
plt.show()
```

## Histogram

**Use when:** Visualizing the distribution of a dataset.

**Explanation:**
Shows how many values fall into ranges (bins).
Great for examining frequency, normal distribution, etc.




```python
data = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5]

plt.hist(data, bins=5, color='coral')
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

## Customize Your Plot


```python
plt.plot(x, y, color='green', marker='o', linestyle='--', linewidth=2)
plt.title("Styled Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

## Multiple Plots in One

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]

plt.plot(x, y1, label='Ascending')
plt.plot(x, y2, label='Descending')
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```

 ## Subplots (Multiple Charts)

**Use when:** You want to show multiple plots in one figure.

**Explanation:**
Arrange multiple plots (rows Ã— columns) in one figure.
Useful for comparing related charts side-by-side.



```python
plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Left Plot")

plt.subplot(1, 2, 2)  # plot 2
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Right Plot")

plt.tight_layout()
plt.show()
```

## Box Plot (Optional but Powerful)

**Use when:** You want to summarize distribution & detect outliers.

**Explanation:**
Shows median, quartiles, min, max, and outliers.
Ideal for statistical summaries.

```python
data = [7, 15, 13, 18, 90, 100, 12, 14]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

| Plot Type    | Use For                             | Best When You Want To...                  |
| ------------ | ----------------------------------- | ----------------------------------------- |
| Line Plot    | Trends or changes over time         | Show progression (e.g., temperature)      |
| Bar Chart    | Categorical comparisons             | Compare values side by side               |
| Pie Chart    | Proportions or parts of a whole     | Show percentage distribution              |
| Scatter Plot | Correlation between two variables   | Detect relationships, clusters            |
| Histogram    | Frequency distribution              | See spread of values, shape of data       |
| Subplots     | Multiple plots in one figure        | Compare different visualizations together |
| Box Plot     | Statistical distribution & outliers | See spread, median, and extreme values    |
